```python
import random

# 모든 trip 정보를 저장할 리스트
trips = []

# 첫 번째 그룹 (W)
for i in range(0, 20):
    depart = random.uniform(0, 200)
    trips.append((depart, f'veh1{i}', "-403444633#2", "998282414#2"))

for i in range(0, 100): # 직진
    depart = random.uniform(0, 200)
    trips.append((depart, f'veh2{i}', "-403444633#2", "480703052#0"))

for i in range(0, 20):
    depart = random.uniform(0, 200)
    trips.append((depart, f'veh3{i}', "-403444633#2", "135763041#3"))

# 두 번째 그룹 (E)
for i in range(0, 10):
    depart = random.uniform(0, 200)
    trips.append((depart, f'e1{i}', "998282428#0", "135763041#3"))

for i in range(0, 60):  # 직진
    depart = random.uniform(0, 200)
    trips.append((depart, f'e2{i}', "998282428#0", "403444633#1"))

for i in range(0, 12):
    depart = random.uniform(0, 200)
    trips.append((depart, f'e3{i}', "998282428#0", "998282414#2"))

# 세 번째 그룹 (S)
for i in range(0, 5):
    depart = random.uniform(0, 200)
    trips.append((depart, f's1{i}', "135763041#0", "480703052#0"))

for i in range(0, 20): # 직진
    depart = random.uniform(0, 200) 
    trips.append((depart, f's2{i}', "135763041#0", "135763041#3"))

for i in range(0, 5):
    depart = random.uniform(0, 200)
    trips.append((depart, f's3{i}', "135763041#0", "403444633#1"))

# 네 번째 그룹 (N)
for i in range(0, 5):
    depart = random.uniform(0, 200)
    trips.append((depart, f'n1{i}', "998282414#0", "403444633#1"))

for i in range(0, 20): # 직진
    depart = random.uniform(0, 200) 
    trips.append((depart, f'n2{i}', "998282414#0", "998282414#2"))

for i in range(0, 5):
    depart = random.uniform(0, 200)
    trips.append((depart, f'n3{i}', "998282414#0", "480703052#0"))

# depart 값을 기준으로 정렬
trips.sort()

# 정렬된 trip 출력
for depart, trip_id, fr, to in trips:
    print(f'<trip id="{trip_id}" type="veh_passenger" depart="{depart:.2f}" departLane="best" departSpeed="max" from="{fr}" to="{to}"/>')
```
